### Data Analysis

* Count of cancer type

```python
plt.figure(figsize=(13, 6.5))
plt.title('Count of cancer type')
data['diagnosis'].value_counts().plot(kind='bar')
```

Number of Malign and Bening is shown above, with over 350 Bening and Malign slightly over 200  diagnosed.

```python
# percentage of diagnosis
pie, ax = plt.subplots(figsize=[13,6.5])
labels=['Benign','Malign']
#labels = data.diagnosis.value_counts(normalize=True).keys()
plt.pie(data['diagnosis'].value_counts(), explode=(0,0.1), labels=labels, autopct="%.1f%%", shadow=True)
plt.title('Percentage of diagnosis', fontsize=14)
plt.legend()
plt.show()
```

Diagnosis Column have 2 unique values Malignant(M) and Benign(B) having count of 62.7 and 37.3 percent respectively.

```python
fig, ax = plt.subplots(2, 2, figsize=(15, 15))
sns.scatterplot(x='fractal_dimension_mean', y='area_mean', hue="diagnosis",
                data=data, ax=ax[0][0], palette='magma')
sns.scatterplot(x='fractal_dimension_worst', y='area_worst', hue="diagnosis",
                data=data, ax=ax[0][1], palette='magma')
sns.scatterplot(x='smoothness_se', y='radius_mean', hue="diagnosis",
                data=data, ax=ax[1][0], palette='magma')
sns.scatterplot(x='symmetry_se', y='radius_worst', hue="diagnosis",
                data=data, ax=ax[1][1], palette='magma');
```

* Heatmap

To find the most correlated features

```python
f, ax = plt.subplots(figsize = ( 20, 20))
sns.heatmap(data.corr(), annot = True, linewidth = 0.5, fmt = '.1f', ax = ax )
```

* Distribution of attributes

```python
# function that plots histogram
def plot_hist(df:pd.DataFrame, column:str, color:str)->None:
    plt.figure(figsize=(9, 7))
    sns.displot(data=df, x=column, color=color, kde=True, height=7, aspect=2)
    plt.title(f'Distribution of {column}', size=20, fontweight='bold')
    plt.show()
```

```python
plot_hist(data, 'radius_mean', 'green')
```

```python
plot_hist(data, 'texture_worst', 'red')
```
